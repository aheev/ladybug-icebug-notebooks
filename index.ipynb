# Ladybug + Icebug Notebook Index

Notebooks are listed from simpler graph-query examples to more advanced storage and analytics workflows.

## Notebooks

1. [Python Tutorial: Analyze a Social Network](ladybug-python-social-network.ipynb)  
   Introductory Ladybug graph querying from Python.

2. [Star Wars Graph Stored in DuckDB](ladybug_duckdb_attached_starwars.ipynb)  
   Attach a DuckDB database and declare a relationship table backed by an attached DuckDB table.

3. [Star Wars Graph Analytics](ladybug_icebug_starwars.ipynb)  
   Query Arrow-backed data, persist it to native Ladybug storage, export CSR, and run Icebug analytics.

4. [Star Wars Analytics with Arrow CSR](ladybug_icebug_memory_starwars.ipynb)  
   Build Arrow CSR tables in memory and share them between Ladybug and Icebug.

5. [Ladybug Icebug Community Detection](ladybug_icebug_comm_detection.ipynb)  
   Export an in-memory Ladybug graph to Parquet, import it into a fresh instance, run Icebug Leiden, and write communities back to the nodes.

6. [Karate Club from icebug-disk](ladybug_icebug_disk_karate.ipynb)  
   Convert DuckDB tables to icebug-disk Parquet, load them natively in Ladybug, and run Icebug algorithms.

7. [pgembed Logical Replication](pgembed_logical_replication.ipynb)  
   Start two embedded Postgres clusters with pgembed over Unix domain sockets, publish logical WAL changes from the primary, and read replicated rows from the secondary.

## Open the Index

```bash
uv run jupyter notebook index.ipynb
```

Open an individual notebook with:

```bash
uv run jupyter notebook <notebook>.ipynb
```

Execute a notebook from the CLI with:

```bash
uv run jupyter nbconvert --to notebook --execute ladybug_icebug_starwars.ipynb --inplace
```

The pgembed replication notebook uses an optional dependency group:

```bash
uv sync --extra postgres
```
